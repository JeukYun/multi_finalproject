```python
import pandas as pd
import numpy as np
```

```python
# 매핑 대상 csv파일 불러오기

main = pd.read_csv("template_수정본.xlsx - 구인구직취업현황_merged.csv")
main = main.iloc[:,:3]
main['사업체수'] = pd.Series() # 매핑해야 할 컬럼 생성
main
```

```python
# 매핑 힐 csv파일 불러오기
map = pd.read_excel("사업장현황(전체)_18~23.xls").reset_index(drop = True)

# 필요한 부분만 슬라이싱
map = map.iloc[1:]
map
```

```python
# "10차_"를 제거하면 main의 산업_대분류 컬럼의 값과 같음 -> 제거
map['산업_대분류'] = map['산업_대분류'].str.replace("10차_","")
map
```

```python
# 마감년월, 시도의 결측치를 이전값으로 채우기
map['마감년월'] = map['마감년월'].fillna(method= 'ffill')
map['(사업장)시도'] = map['(사업장)시도'].fillna(method= 'ffill')
map
```

```python
# 산업대분류 결측행 제거
map['산업_대분류'] = map['산업_대분류'].dropna()
map
```

```python
map['마감년월'] = map['마감년월'].str.replace("년","").str.replace("월","")
```

```python
# 결측 제거
map = map.dropna()
```

```python
# unique값 확인
map['(사업장)시도'].unique()
```

```python
#확인
map
```

```python
# main의 (사업장)시도의 값과 같도록 변경
map['(사업장)시도'].loc[map['(사업장)시도'] == '경상북도'] = '경북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '경상남도'] = '경남'
map['(사업장)시도'].loc[map['(사업장)시도'] == '전라북도'] = '전북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '전라남도'] = '전남'
map['(사업장)시도'].loc[map['(사업장)시도'] == '충청북도'] = '충북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '충청남도'] = '충남'
map['(사업장)시도'] = map['(사업장)시도'].str[:2]
```

```python
# 변경 확인
map['(사업장)시도'].unique()
```

```python
map.columns = ['년/월', '시도', '산업_대분류', '사업체수']
```

```python
main['년/월'] = main['년/월'].astype('object')
```

```python
key = ['년/월', '시도', '산업_대분류']
merge_df = pd.merge(main, map, left_on = key, right_on = key, how = 'left' )
```

```python
main['사업체수'] = 0
```

```python
main['년/월'] = main['년/월'].astype(str)
```

```python
merged_df = main.merge(map, on=['년/월', '시도', '산업_대분류'], how='left')
merged_df
```

```python
merged_df['사업체수_y'] = merged_df['사업체수_y'].fillna(0).astype('int')
```

```python
merged_df.to_csv("merge.csv", encoding = 'ANSI')
```
