직종_중분류 - 정보통신 관련직

산업_대분류 - 09차_출판, 영상, 방송통신 및 정보서비스업 -> '정보통신업', '제조업', '도매 및 소매업', '전문, 과학 및 기술 서비스업', '사업시설 관리, 사업 지원 및 임대 서비스업'

```python
def pre_df(df, temp):
    df.fillna(method='pad', inplace=True)
    df.drop('구직건수(월)', axis=1, inplace=True)
    df['마감년월'] = df['마감년월'].str.replace('년 ', '', regex=True).replace('월', '', regex=True)
    df['산업_대분류'] = df['산업_대분류'].str.replace('09차_', '', regex=True)
    df['직종_중분류'] = df['직종_중분류'].str.replace('2007직종_', '', regex=True)
    
    df.loc[df['산업_대분류'] == '출판, 영상, 방송통신 및 정보서비스업', '산업_대분류'] = '정보통신업'
    df = df.loc[df['직종_중분류'] == '정보통신 관련직']
    df.drop('직종_중분류', axis=1, inplace=True)
    df.columns = temp.columns
    
    return df


def merge_file(year):
    df = pd.read_excel(f'data/구인구직취업현황(월)_{year}.xlsx', engine='openpyxl')
    temp = pd.read_excel('data/template_2018이전.xlsx', engine='openpyxl')
    new_temp = temp.iloc[:, :5]

    import warnings

    warnings.filterwarnings('ignore')

    df = pre_df(df, new_temp)

    new_temp['년/월'] = new_temp['년/월'].astype('str')
    new_temp['구인인원'].fillna(0, inplace=True)
    new_temp['취업건수'].fillna(0, inplace=True)

    new_temp['구인인원'] = new_temp['구인인원'].astype('int64')
    new_temp['취업건수'] = new_temp['취업건수'].astype('int64')

    temp_year = new_temp.loc[new_temp['년/월'].str.startswith(f'{year}')]

    merged = pd.merge(temp_year, df, how='left', on=['년/월', '시도', '산업_대분류'])

    merged.drop(columns=['구인인원_x', '취업건수_x'], inplace=True)

    merged.columns = new_temp.columns

    merged.fillna(0, inplace=True)

    merged.to_csv(f'data/구인구직취업현황_{year}_merged.csv', index=False)
```

```python
merge_file(2016)
merge_file(2017)
```
