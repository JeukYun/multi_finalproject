```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/NCS_직무기준_직업표.csv', encoding='cp949')
data
```

```python
data.columns = ['NCS_대분류', 'NCS_중분류', 'NCS_소분류', 'NCS_세분류', 'KECO_코드', 'KECO_세분류']
data.drop(index=0, inplace=True)
data
```

```python
import re
first_num = re.sub(r'[^0-9]', '', data.iloc[0, 0])
print(first_num, type(first_num))
second_num = re.sub(r'[^0-9]', '', data.iloc[0, 0])
print(second_num, type(second_num))
third_num = re.sub(r'[^0-9]', '', data.iloc[0, 0])
print(third_num, type(third_num))
forth_num = re.sub(r'[^0-9]', '', data.iloc[0, 0])
print(forth_num, type(forth_num))
```

```python
import re

ncs_code = []

for i in range(1759):
    first_num = re.sub(r'[^0-9]', '', data.iloc[i, 0])
    if len(first_num) == 1:
        first_num = '0' + first_num
    second_num = re.sub(r'[^0-9]', '', data.iloc[i, 1])
    if len(second_num) == 1:
        second_num = '0' + second_num
    third_num = re.sub(r'[^0-9]', '', data.iloc[i, 2])
    if len(third_num) == 1:
        third_num = '0' + third_num
    forth_num = re.sub(r'[^0-9]', '', data.iloc[i, 3])
    if len(forth_num) == 1:
        forth_num = '0' + forth_num
    
    ncs_code.append(first_num + second_num + third_num + forth_num)

data['NCS_Code'] = ncs_code
data.head()
```

```python
job_code = data[['NCS_Code', 'KECO_코드', 'KECO_세분류']]
job_code
```

```python
ncs_info = data[['NCS_Code', 'NCS_대분류', 'NCS_중분류', 'NCS_소분류', 'NCS_세분류']]
ncs_info.drop_duplicates(subset='NCS_Code', inplace=True)
ncs_info.reset_index(drop=True, inplace=True)
ncs_info
```

```python
job_code.to_csv('data/NCS_code_jobname.csv', index=False, encoding='utf-8')
ncs_info.to_csv('data/NCS_code_info.csv', index=False, encoding='utf-8')
```
