- 태그 의미
    - address: 주소
    - courseMan: 수강비
    - instCd: 훈련기관 코드
    - ncsCd: NCS 직무 코드
    - realMan: 실제 훈련비
    - regCourseMan: 수강신청 인원
    - subTitle: 부 제목 -> 훈련기관명
    - subTitleLink: 부 제목 링크
    - telNo: 훈련기관 전화번호
    - title: 제목
    - titleIcon: 제목 아이콘
    - titleLink: 제목 링크
    - traEndDate: 훈련 종료일자
    - traStartDate: 훈련 시작일자
    - trainTarget: 훈련대상
    - trainTargetCd: 훈련구분
    - trainstCstId: 훈련ID
    - trngAreaCd: 지역코드(중분류)
    - trprDegr: 훈련과정 순차
    - trprId: 훈련과정 ID
    - yardMan: 정원
