```python
import pandas as pd
import numpy as np
```

```python
# 매핑 대상 csv파일 불러오기

main = pd.read_excel(r"C:\Data_Engineer_Study\멀캠_PJT_2\multi_finalproject\data\Mapping_data\template_수정본.xlsx")
main = main.iloc[:,:3]
main['사업체수'] = pd.Series() # 매핑해야 할 컬럼 생성
main
```

```python
# 매핑 힐 csv파일 불러오기
map18 = pd.read_excel(r"C:\Data_Engineer_Study\멀캠_PJT_2\multi_finalproject\data\Mapping_data\사업장현황(전체)_18~23.xls").reset_index(drop = True)

# 필요한 부분만 슬라이싱
map18 = map18.iloc[1:]
map18
```

```python
# 매핑 힐 csv파일 불러오기
map13 = pd.read_excel(r"C:\Data_Engineer_Study\멀캠_PJT_2\multi_finalproject\data\Mapping_data\사업장현황(전체)_13~17.xls").reset_index(drop = True)

# 필요한 부분만 슬라이싱
map13 = map13.iloc[1:]
map13
```

```python
map = pd.concat([map13, map18], axis = 0)
```

```python
map
```

```python
# 산업_대분류 컬럼의 "10차_", "9차_" 제거
map['산업_대분류'] = map['산업_대분류'].str.replace("10차_","")
map['산업_대분류'] = map['산업_대분류'].str.replace("09차_","")
map
```

```python
# 마감년월, 시도의 결측치를 이전값으로 채우기
map['마감년월'] = map['마감년월'].fillna(method= 'ffill')
map['(사업장)시도'] = map['(사업장)시도'].fillna(method= 'ffill')
map
```

```python
map['마감년월'] = map['마감년월'].str.replace("년","").str.replace("월","")
```

```python
# 결측 제거
map = map.dropna()
```

```python
# unique값 확인
map['(사업장)시도'].unique()
```

```python
# main의 (사업장)시도의 값과 같도록 변경
map['(사업장)시도'].loc[map['(사업장)시도'] == '경상북도'] = '경북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '경상남도'] = '경남'
map['(사업장)시도'].loc[map['(사업장)시도'] == '전라북도'] = '전북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '전라남도'] = '전남'
map['(사업장)시도'].loc[map['(사업장)시도'] == '충청북도'] = '충북'
map['(사업장)시도'].loc[map['(사업장)시도'] == '충청남도'] = '충남'
map['(사업장)시도'] = map['(사업장)시도'].str[:2]
```

```python
# 변경 확인
map['(사업장)시도'].unique()
```

```python
map['산업_대분류'].loc[map['산업_대분류'] == '출판, 영상, 방송통신 및 정보서비스업'] = '정보통신업'
map['산업_대분류'].loc[map['산업_대분류'] == '사업시설관리 및 사업지원 서비스업'] = '사업시설 관리, 사업 지원 및 임대 서비스업'
```

```python
job_list = ['정보통신업', '제조업', '도매 및 소매업', '전문, 과학 및 기술 서비스업', '사업시설 관리, 사업 지원 및 임대 서비스업']
map = map.loc[map['산업_대분류'].isin(job_list)].reset_index(drop = True)
map
```

```python
# 5개 산업대분류 추출 완료
map['산업_대분류'].unique()
```

```python
# map 데이터프레임과 main 데이터프레임의 컬럼명 일치시키기
map.columns = main.columns
```

```python
main['년/월'] = main['년/월'].astype(str)
```

```python
main['사업체수'] = 0
main
```

```python
merged_df = main.merge(map, on=['년/월', '시도', '산업_대분류'], how='left')
merged_df
```

```python
merged_df['사업체수_y'] = merged_df['사업체수_y'].fillna(0).astype('int')
```

```python
merged_df.to_csv("merge.csv", encoding = 'ANSI')
```
