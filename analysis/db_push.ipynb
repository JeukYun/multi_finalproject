```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import openpyxl

pymysql.install_as_MySQLdb()
import MySQLdb
```

```python
engine = create_engine("mysql+mysqldb://root:1234@localhost:3306/multi")
conn = engine.connect()
```

```python
ncs_code_info = pd.read_csv('data/NCS_code_info.csv', encoding='cp949')
ncs_code_info
```

```python
ncs_code_info.columns = ['NCS_code', 'code1', 'code2', 'code3', 'code4']
```

```python
ncs_code_info.info()
```

```python
ncs_code_info.to_sql(name='NCS_code_info', con=engine, if_exists='append', index=False)
```
