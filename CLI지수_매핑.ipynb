```python
import pandas as pd
```

```python
cli = pd.read_csv("KOR_CLI.csv")
```

```python
cli = cli[['TIME', 'Value']]
cli.head()
```

```python
def format_time(time_str):
    # 월의 약어를 숫자로 매핑하는 딕셔너리 생성
    month_mapping = {
        'Jan': '01',
        'Feb': '02',
        'Mar': '03',
        'Apr': '04',
        'May': '05',
        'Jun': '06',
        'Jul': '07',
        'Aug': '08',
        'Sep': '09',
        'Oct': '10',
        'Nov': '11',
        'Dec': '12'
    }
    
    # TIME 컬럼의 문자열을 분해하여 연도와 월을 추출
    year = '20' + time_str[:2]  # 연도는 앞 두 자리
    month = month_mapping[time_str[-3:]]  # 월은 마지막 세 자리
    
    # 연도와 월을 결합하여 원하는 형식으로 반환
    formatted_time = year + month
    
    return formatted_time
```

```python
cli['TIME'] = cli['TIME'].apply(format_time)
```

```python
# 매핑 대상 csv 불러오기
main = pd.read_csv("template_수정본.xlsx - 구인구직취업현황_merged.csv")
```
