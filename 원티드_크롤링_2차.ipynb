```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# 크롤링 결과 데이터 받을 DataFrame 생성
wanted_DBA = pd.DataFrame(columns=['stack', 'preference', 'region'], index=[0])

# selenium 실행
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 원티드 개발직군 - DBA + 데이터엔지니어
driver.get('https://www.wanted.co.kr/wdlist/518?country=kr&job_sort=job.latest_order&years=-1&selected=655&selected=10231&locations=all')

# 현재 페이지 높이 = last_height
last_height = driver.execute_script("return document.body.scrollHeight")

PAUSE_SEC = 3

while True:
    # 스크롤 끝까지 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # 스크롤 내린 후 페이지 로딩을 위한 시간
    time.sleep(PAUSE_SEC)
    
    # 스크롤 내린 후 페이지 높이
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # 더이상 스크롤이 내려가지 않으면 스크롤 내리는 반복문 멈춤
    if new_height == last_height:
        break
    # 스크롤 내린 후 페이지 높이를 현재 페이지 높이 변수에 저장
    last_height = new_height

time.sleep(3)

position_link = driver.find_elements(By.CSS_SELECTOR, "div.List_List_container__JnQMS ul li div a")
time.sleep(2)

links = []
for i in position_link:
    # 링크 받아오기
    links.append(i.get_attribute('href'))

for i in links:
    # 변수 초기화
    preference = ''
    stack = ''
    # 오류 발생시 넘기도록 설정
    try:
        time.sleep(1)
        driver.get(i)
        time.sleep(5)
        
        # 회사명 받아오기
        company = driver.find_element(By.CSS_SELECTOR, "section.JobHeader_className__HttDA div").text
        # 상세 정보 데이터 받아오기
        detail = driver.find_elements(By.CSS_SELECTOR, "section.JobDescription_JobDescription__VWfcb p")
        del detail[0]
        detail_head = driver.find_elements(By.CSS_SELECTOR, "section.JobDescription_JobDescription__VWfcb h6")
        # 지역정보 받아오기
        region = driver.find_elements(By.CSS_SELECTOR, "span.JobHeader_pcLocationContainer__xRwIv")[0].text
        
        # 필요로하는 우대사항과 스택 데이터만 저장
        for j in range(len(detail)):
            if detail_head[j].text == '우대사항':
                preference = detail[j].text
            elif detail_head[j].text == '기술스택 ・ 툴':
                stack = detail[j].text
        
        target = pd.DataFrame({'company':company, 'stack':stack, 'preference':preference, 'region':region, 'link':i}, index=[0])
        
        wanted_DBA = pd.concat([wanted_DBA, target], ignore_index=True)
    except:
        pass

driver.quit()

wanted_DBA.to_csv('data/wanted_crawling_DBA_DE_pre.csv', index=False)
```
